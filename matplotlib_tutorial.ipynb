```python
plt.subplot(2,2,1)
plt.plot(x,y,'r-')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'b')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title('matplotlib demo')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.plot(x,y)
```

```python
np.pi
```

```python
x=np.arange(0,4*np.pi,0.1)     
y=np.sin(x)
plt.plot(x,y)
```

```python
x=np.arange(0,5*np.pi,0.1)  
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.plot(x,y_sin)
plt.plot(x,y_cos)
```

```python
###  pie chart
labels='pyhton','c++','ruby','java'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,shadow=True)
```
